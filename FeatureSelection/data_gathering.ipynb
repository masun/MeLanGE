# Data import

```python
df = pd.read_csv("Cog_counts_wt_metadata.csv")
df.head()
```

```python
df_filt = df[(df["Genus"]=="Aquimarina") | (df["Genus"] == "Tenacibaculum")]
df_filt = df_filt.loc[:, (df_filt != 0).any(axis=0)]
df_filt.head()
```

```python
df_filt = df_filt.drop(columns=["Origin", "MAG", "Family"]).set_index("Assembly accession")
df_filt.head()
```
